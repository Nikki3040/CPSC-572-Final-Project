```python
# # use networkx to create a network of nodes and edges from csv file that is imported
# from the database
!pip install pandas
!pip install scipy
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import csv
import numpy as np
import os

# import the csv file called Master.csv from local directory
# this is the file that is created by the database
# it is a csv file with the following columns to be used as nodes:
# source airport, destination airport
# where source airport and destination airport are the nodes that are connected
# and weight is the weight of the edge between the two nodes
def import_csv():
    # get the current working directory
    cwd = os.getcwd()
    # get the file name
    file_name = "Master.csv"
    # get the full path to the file
    file_path = os.path.join(cwd, file_name)
    # read the csv file
    df = pd.read_csv(file_path)
    # return the dataframe  
    return df

# create a networkx graph
# add nodes for Source City and Destination City
# add edges between them
def create_graph(df):
    # create a graph
    G = nx.Graph()
    # add nodes to the graph
    G.add_nodes_from(df['Source City'])
    G.add_nodes_from(df['Destination City'])
    # add edges to the graph
    G.add_edges_from(zip(df['Source City'], df['Destination City']))
    # return the graph
    return G

# draw the graph
def draw_graph(G):
    # draw the graph
    pos = nx.spring_layout(G, k=0.0000150, iterations=20)
    nx.draw(G, pos, with_labels=True)
    # show the graph
    plt.draw()
    plt.show()

# main function
def main():
    # import the csv file
    df = import_csv()
    # create a graph
    G = create_graph(df)
    # draw the graph
    draw_graph(G)
```
